```python
import pandas as pd
from scipy import sparse
```

```python
ratings = pd.read_csv('dataset/ratings.csv')
ratings
```

```python
movies = pd.read_csv('dataset/movies.csv')
movies
```

```python
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)
userRatings.head()
```

```python
userRatings.shape
```

```python
item_similarity_df=userRatings.corr(method='pearson')
item_similarity_df
```

```python
def get_similar(movie_name,rating):
    similar_score = item_similarity_df[movie_name]*(rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
```

```python
action_lover = [("Zack and Miri Make a Porno (2008)",5),("(500) Days of Summer (2009)",3),("Zero Effect (1998)",1)]
similar_scores = pd.DataFrame()
for movie,rating in action_lover:
    similar_scores = similar_scores.append(get_similar(movie,rating))
similar_scores
```
